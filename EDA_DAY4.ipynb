# Map

1. it tries to link the data basis some col values

2. Syntax - map(function_obj,variable)

```python
#Taking city as reference map season in the in the humidity dataset
Season ={"new york":"Winter","mumbai":"Rainy","beijing" : "Spring"}
Season
#the keyvalues should be exactly same as column on whose basis we are goin to map as it will not map the city name if we change case or name it will not able to find it in the table
```

```python
#Map(link) the season column in the humidity data taking city as reference
Humid["Season"]=Humid["city"].map(Season)
Humid
```

```python
#traffic keyvalue pair - Dictionary
Traffic = {"new york": "High traffic","beijing" : "low traffic", "mumbai":"Disastrous traffic"}
```

```python
#map the traffic according to city
Humid["Traffic"]=Humid["city"].map(Traffic)
Humid
```
